# Write CSV

```python
import pandas as pd

dis = {"a":[1,2,3,4,5,6],"s":[1,2,3,4,5,6],"d":[1,2,3,4,5,6]}

d = pd.DataFrame(dis)

print(d)

d.to_csv("Test_new.csv")
```

# Read CSV Files in Python Pandas

```python
import pandas as pd

csv_1 = pd.read_csv("//Users//princemaurya//Desktop//Jupyter//stock_data.csv")

csv_1
```

Now ab hm csv file ke andar koi bhi work/function perform kr skte hai

if we want to get any row then use nrows function

```python
csv_2 = pd.read_csv("//Users//princemaurya//Desktop//Jupyter//stock_data.csv",nrows=3)

csv_2
# print(type(csv_2))
```

if we want to get any column then use usecols function

```python
csv_3 = pd.read_csv("//Users//princemaurya//Desktop//Jupyter//stock_data.csv",usecols=["SYMBOL","CHANGE"])

csv_3
```

Also we can pass the index number of the column to get the column

```python
csv_3 = pd.read_csv("//Users//princemaurya//Desktop//Jupyter//stock_data.csv",usecols=[0,3])

csv_3
```

if we want to remove any row than use skiprows function

```python
csv_3 = pd.read_csv("//Users//princemaurya//Desktop//Jupyter//stock_data.csv",skiprows=[0,3])

csv_3
```

agar hme kisi particular column ko as a index use karna hai than use index_col functions

```python
csv_3 = pd.read_csv("//Users//princemaurya//Desktop//Jupyter//stock_data.csv",index_col="SYMBOL")

csv_3
```

Agar hme given row me se kisi ek ko header bnana hai than use header function

```python
csv_3 = pd.read_csv("//Users//princemaurya//Desktop//Jupyter//stock_data.csv",header = 2)

csv_3
```

agar hm name parameter ka use krte hai to hm heading ko bhi change kr skte hai

```python
csv_3 = pd.read_csv("//Users//princemaurya//Desktop//Jupyter//stock_data.csv",names=["col1","col2","col3","col4","col5"])

csv_3
```

agar hm header ko hta ke new col dalna chahte hai then use header and prefix function

```python
import pandas as pd

# Read CSV without header
csv_1 = pd.read_csv("/Users/princemaurya/Desktop/Jupyter/stock_data.csv", header=None)

# Assign column names manually with a prefix
csv_1.columns = [f"col{i}" for i in range(csv_1.shape[1])]

csv_1
```
