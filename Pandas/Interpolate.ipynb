# Interpolate

```python
import pandas as pd

var = pd.read_csv("//Users//princemaurya//Desktop//Jupyter//Interpolate.csv")

var
```

interpolate function is used to fill the the data automatically 

```python
var.interpolate()
```

Note - Interpolate function only works on the number.(It doesn't works on the alphabet)

```python
var.interpolate("linear")
```

```python
var.interpolate("zero")
```

```python
var.interpolate("linear",axis=0)
```

above code column/column data ko fill karne ke kaam aata hai

now agar hme row/row data ko fill krna hai then

```python
# var.interpolate("linear",axis=1) agar axis=1 ka use kr rhe hai to use case me datatype sare row and column ka same hona chahiye otherwise it will give an error.
```

agar hm kuchh ki data ko fill karna chahte hai than us case me hm limit parameter ka use karte hai

```python
var.interpolate(limit=2)
```

Note - hm limit ki direction ko bhi de skte hai forword,backword or both etc.

```python
var.interpolate(limit_direction="forward",limit=2)
```

```python
var.interpolate(limit_direction="backward",limit=2)
```

```python
var.interpolate(limit_direction="both",limit=2)
```

limit_area ye hme btata hai ki interpolate use karna hai ya nhi,hm uska area specify kr skte hai.

```python
var.interpolate(limit_area="inside")
```

above inside limit ke andar jitne bhi blank data hai usko fill karke de dega.

and outside ka use karne se ye sari NaN value ko rakhata hai

```python
var.interpolate(limit_area="outside")
```

inplace - agar hm apne data ko original data ke andar place karana chahte hai to us case me inplace function ka use karte hai.

and agar hme multiple copy banani hoti hai to inplace ko False karte hai

```python
var.interpolate(limit_direction="both",limit=2,inplace=True)
```
