# fillna

```python
import pandas as pd

var = pd.read_csv("//Users//princemaurya//Desktop//Jupyter//company_data_row22_cleared.csv")

var
```

```python
var.fillna("python")
```

agar hm kisi particular column ke andar kisi data ko fill karna chahte hai then use dictionary

```python
var.fillna({"Code":"python","Company":"c","Value C":""	})
```

agar hm null value ko farword ya backword data se fill krna chahte hai then use

```python
var.fillna(method="ffill")
```

for the backword filling

```python
var.fillna(method="bfill")
```

agar hm kisi data ko uske axis ke along fill krna chahte hai then use

```python
import pandas as pd

# enable future behavior explicitly
pd.set_option('future.no_silent_downcasting', True)

# assuming var is already defined as a DataFrame:
var = var.bfill(axis=1)  # backward fill along columns
```

```python
var.fillna(method="ffill",axis=1)
```

on the above code aage wale data picche wale me fill ho chuka hai.
