# Melt Function

```python
import pandas as pd

var = pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})

var
```

above table horizontal chal rhi hai, but ab hm is table ko vertically chalana chahte hai to use melt(var)

```python
pd.melt(var)
```

agar hm kisi id ko variable me use karna chahate hai to 

```python
pd.melt(var,id_vars=["days"])
```

agar hm variable ke name ko change krna hai than use 

```python
pd.melt(var,id_vars=["days"],var_name="python")
```

agar hm value ka name change krna chahte hai than use

```python
pd.melt(var,id_vars=["days"],var_name="python",value_name="wscube")
```

# Pivot Function

```python
import pandas as pd

var = pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})

var
```

agar hm chahte hai is table ke andar data student name ke according set ho jaye than use pivot

```python
var.pivot(index="days",columns="st_name")
```

```python
import pandas as pd

var = pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","b","a","a","b","b"],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})

var
```

```python
var.pivot(index="days",columns="st_name")
```

agar hm chahte hai ki inme se keval eng ka data mile maths ka nhi than use

```python
var.pivot(index="days",columns="st_name",values="eng")
```

so is trah se hm kisi ek particular column ka data get kr skte hai

```python
import pandas as pd

var = pd.DataFrame({"days":[1,1,1,1,2,2],"st_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})

var
```

Now is table ka arrangement hme krna hai proper way me than use pivot_table & isme hm mathematical operation bhi perform kr skte hai using aggfunc function

```python
var.pivot_table(index="st_name",columns="days",aggfunc="mean")
```

for sum

```python
var.pivot_table(index="st_name",columns="days",aggfunc="sum")
```

margins = True karne pr hme iska avg ka sum mil jata hai.

```python
var.pivot_table(index="st_name",columns="days",aggfunc="sum",margins="True")
```

Now for calculating the mean of avg use

```python
var.pivot_table(index="st_name",columns="days",aggfunc="mean",margins="True")
```
