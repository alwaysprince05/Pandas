inplace function: agar hmne kisi data ko replace kiya to vo us data ki copy nhi bnayega, ye original data ke andar hi replace kr deta

```python
import pandas as pd

var = pd.read_csv("//Users//princemaurya//Desktop//Jupyter//stocks_data2.csv")

var
```

```python
import pandas as pd
import numpy as np

# Step 1: Replace 1s with NaN
var_replaced = var.replace(1, np.nan)

# Step 2: Forward fill (only NaNs)
var_filled = var_replaced.ffill(limit=3,inplace=True)

# Step 3: Update original DataFrame #chatgpt
var.update(var_filled)

print(var)
```

```python
var.replace(1,method="ffill",limit=3,inplace=True) # Sir method
```
